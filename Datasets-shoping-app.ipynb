```python
# Identify missing values
print(df.isnull().sum())
```

```python
df = df.dropna(subset=['seller_name', 'seller_information'])
```

```python
# Fill numeric missing values

df['discount'] = df['discount'].fillna(0)

# Fill text missing values

df['what_customers_said'] = df['what_customers_said'].fillna('No Review')

df['videos'] = df['videos'].fillna('No Video')

df['variations'] = df['variations'].fillna('No Variation')
```

```python
print(df.isnull().sum())
```

```python
# Convert to string

df['final_price'] = df['final_price'].astype(str)

# Remove currency symbols and commas

df['final_price'] = df['final_price'].str.replace(r'[^0-9.]', '', regex=True)

# Convert to numeric

df['final_price'] = pd.to_numeric(df['final_price'], errors='coerce')
```

```python
# Remove rows where final_price is NaN

df = df.dropna(subset=['final_price'])
```

```python
print(df['final_price'].head())
```

```python
# Remove duplicate records

df = df.drop_duplicates()

# Check dataset shape after removing duplicates
print(df.shape)
```

```python
filtered_df = df[df['final_price'] > 500]

print(filtered_df.head())
```

```python
selected_columns = df[['title', 'final_price', 'category']]

print(selected_columns.head())
```

```python
# Create quantity column

df['quantity'] = 1

# Create total_amount column

df['total_amount'] = df['final_price'] * df['quantity']

# Display new columns
print(df[['final_price', 'quantity', 'total_amount']].head())
```

```python
# Save cleaned dataset

df.to_csv('cleaned_dataset.csv', index=False)

print("Cleaned dataset saved successfully!")
```
